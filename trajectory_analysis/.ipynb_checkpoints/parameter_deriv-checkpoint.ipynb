# Parameter Derivation and Trajectory Visualization

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import os,glob
import preprocess as prc
import seaborn as sns; sns.set_style('darkgrid'); sns.set_context('notebook')
```

Import trial file.

```python
ind = 2
trial = 1
```

```python
tr = pd.read_csv("labelled_data/individual_"+str(ind)+"/trial_"+str(trial)+".csv")
fig_file = "labelled_data/individual_"+str(ind)+"/plots_i"+str(ind)+"/trial_"+str(trial)
```

## Visualizing a Trajectory

Visualize data and get jump parameters.

```python
turn_ang, azimuth, approach_angle, take_off_angle, vel = prc.preprocess(tr.copy(), fig_file, zoom = True)
```

Construct dataframe by parsing through all trials.

```python
ind = [1,2,3]
df = pd.DataFrame(np.zeros((13,6)),columns = {'turn_ang','azimuth','approach_ang','take_off_ang','vel','ind'})
df = df[['turn_ang', 'azimuth', 'approach_ang', 'take_off_ang', 'vel', 'ind']]
```

```python
count = 0
for i in ind:
    folder_path = 'labelled_data/individual_'+str(i)+'/'
    for filename in glob.glob(os.path.join(folder_path, '*.csv')):
        tr = pd.read_csv(filename)
        trial = filename[-5]
        fig_file = "labelled_data/individual_"+str(i)+"/plots_i"+str(i)+"/trial_"+trial
        turn_ang, azimuth, approach_angle, take_off_angle, vel = prc.preprocess(tr.copy(), 
                                                                                fig_file, zoom = True)
        df.iloc[count,:] = np.array([turn_ang, azimuth, approach_angle, take_off_angle, vel, i])
        count += 1
```

Write to csv file.

```python
pd.DataFrame.to_csv(df,"parameters.csv",index = False)
```
