# Visualizing Parameter Distributions

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set_style('darkgrid'); sns.set_context('notebook')
import circular as crc
```

```python
df = pd.read_csv("parameters.csv")
```

```python
dist_file = "dist_figs/"
```

### 1\. Take-Off Velocity

```python
sns.distplot(df['vel'][df['ind'] == 1], hist = False, rug = True, kde = False)
sns.distplot(df['vel'][df['ind'] == 2], hist = False, rug = True, kde = False)
sns.distplot(df['vel'][df['ind'] == 3], hist = False, rug = True, kde = False)
sns.distplot(df['vel'], hist = False, color = 'purple')
plt.legend(['$I_1$', '$I_2$', '$I_3$'])
plt.xlabel('$take-off\ velocity\ (cm/s)$');
plt.savefig(dist_file+"Take off Velocity",dpi = 300,bbox_inches = 'tight'); 
```

### 2\. Take-Off Angle

Plot take off angle between $0^\circ$ and $90^\circ$

```python
a1 = np.array(df['take_off_ang'][df['ind']==1])
a2 = np.array(df['take_off_ang'][df['ind']==2])
a3 = np.array(df['take_off_ang'][df['ind']==3])
crc.circ_dist_plotter(a1,a2,a3,8,4.2,"E",dist_file,"Take Off Angles",0,0.8,0,90)
#circ_dist_plotter(a1,a2,a3,bin_size,radius,zero_direction,fig_file,title,start_ang,bottom,min_ax_ang,max_ax_ang)
```

### 3\. Turning Angle

$0^\circ$ represents initial body axis before turning.

```python
a1 = np.array(df['turn_ang'][df['ind']==1])
a2 = np.array(df['turn_ang'][df['ind']==2])
a3 = np.array(df['turn_ang'][df['ind']==3])
crc.circ_dist_plotter(a1,a2,a3,8,8.7,"N",dist_file,"Turning Angles",-180,1.8,0,360)
```
