```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set_style('dark'); sns.set_context('notebook')
```

```python
def plotter(degrees,dist):
    
    bin_size = 10
    a , b=np.histogram(degrees, bins=np.arange(0, 360+bin_size, bin_size))
    centers = np.deg2rad(np.ediff1d(b)//2 + b[:-1])

    fig = plt.figure(figsize = (5,5))
    ax = fig.add_subplot(111, projection='polar')
    ax.bar(centers, a, width=np.deg2rad(bin_size), bottom=15.0, color='crimson', edgecolor='crimson', alpha = 0.5)
    ax.set_theta_zero_location("S")
    ax.set_yticklabels([])

    plt.title('$'+dist+'\ distribution$', pad = 20)

    plt.savefig('th_escape-angle_'+dist,dpi = 300); plt.show()
```

```python
sigma = 30; mu = 180
degrees = sigma * np.random.randn(500) + mu
plotter(degrees, 'normal')
```

```python
mu1 = 150; mu2 = 180; mu3 = 210
sigma1 = 9; sigma2 = 9; sigma3 = 9
degrees = [(sigma1*np.random.randn(160) + mu1),(sigma2*np.random.randn(160) + mu2),(sigma3*np.random.randn(160) + mu3)]
plotter(degrees, 'multimodal')
```

```python
b = 140; a = 60; w = b-a
degrees = w*np.random.rand(500) + b
plotter(degrees, 'uniform')
```
